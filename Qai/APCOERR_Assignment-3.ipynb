This is one of the interesting algorithm, we will try to build the circuit and understand, Let us go here!!!!!!

```python
from qiskit import *
```

```python
qc=QuantumCircuit(3,3) #building quatum circuit with 3 quantum and 3 classical registers
```

```python
qc.draw("mpl")
```

```python
qc.x(0)
qc.barrier()
qc.draw("mpl")
```

```python
qc.h(1)
qc.cx(1,2)
qc.draw(output="mpl")
```

```python
qc.cx(0,1)
qc.h(0)
qc.draw(output="mpl")
```

```python
qc.barrier()
qc.measure([0,1],[0,1])
qc.draw(output="mpl")
```

```python
qc.barrier()
qc.cx(1,2)
qc.cz(0,2)
qc.draw(output="mpl")
```

Now this is complete circuit of quantum teleportations, where $q_0$ is transferred to the $q_2$. you can see this through measurement.


As $q_0$ contains 1 same value will be there in $q_2$ let us check

```python
qc.measure(2,2)
simulator=Aer.get_backend('qasm_simulator')
```

```python
result=execute(qc, backend=simulator,shots=1000).result()
```

```python
counts=result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(counts)
```

```python
print(counts)
```

From the above result it is clear that in every measurement of $c_2$, classical bit 2 contains all the time 1, it means we teleported $q_0$ to $q_2$
